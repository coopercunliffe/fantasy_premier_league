```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('top500histories.csv', index_col=0)
df500 = pd.read_csv('top500.csv', index_col=0)
dfteams = pd.read_csv('teams', index_col=0)


```

```python
now_gw = max(df.gameweek)
```

```python
top10 = df.iloc[:now_gw*6]
sns.set()
plt.figure(figsize=(24, 6))
plt.title('Points Each Week')
sns.lineplot(data=top10, x='gameweek', y='total_points', hue='web_name', markers=True)
plt.show()
```

```python
top10 = df.iloc[:now_gw*6]
sns.set()
plt.figure(figsize=(24, 6))
plt.title('Accumulated Points')
sns.lineplot(data=top10, x='gameweek', y='sum_points', hue='web_name')
plt.show()
```

```python
# The 15 players that were leading in week 3 plotted through current week

x = 3

week_x_leaders = df[df['gameweek'] == x].sort_values('sum_points', ascending=False)['element'][:15]
week_x_df = df[df['element'].isin(week_x_leaders)]
sns.set()
plt.figure(figsize=(24, 6))
plt.title('Week X Points Leaders')
sns.lineplot(data=week_x_df, x='gameweek', y='sum_points', hue='web_name')
plt.show()
```
